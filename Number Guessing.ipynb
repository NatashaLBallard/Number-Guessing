<p style="font-size: 16px">
Design a program that randomly chooses a number between any set range. Give users a hint to guess the number. Every time the user guesses wrong, they get another clue, but the score gets reduced. The clue can be multiples, divisible, greater or smaller, or a combination of all.
</p>

<br/>

```python
import random as rand
firstNum = 10
lastNum = 20
compNum = rand.randint(firstNum,lastNum)
print("Computer Number: " + str(compNum))          # For Testing Purposes


def guessNum(userGuess):
    counter = 3
    
    while counter > 0:
        
        if userGuess != compNum :
            # Clue 1 - Even or Odd
            if userGuess % 2 == 0 and compNum % 2 == 0:
                print(str(userGuess) + " is incorrect, but you're getting closer. Clue 1: The number is even.")
                counter -= 1
            elif userGuess % 2 != 0 and compNum % 2 != 0:
                print(str(userGuess) + " is incorrect, but you're getting closer. Clue 1: The number is odd.")
                counter -= 1
            elif userGuess % 2 != 0 and compNum % 2 == 0:
                print(str(userGuess) + " is incorrect, but try again. Clue 1: The number is even.")
                counter -= 1
            else:
                print(str(userGuess) + " is incorrect, but try again. Clue 1: The number is odd.")
                counter -= 1        
                
            # Clue 2 - Divisibility by 3    
            userGuess = int(input("Your guess: "))
            if userGuess == compNum :
                print("Awesome guess. You win!")
                break
            elif compNum % 3 == 0:
                print("Clue 2: The number is divisible by 3.")
                counter -= 1
            else :
                print("Clue 2: The number is not divisible by 3.")
                counter -= 1       
            
            # Clue 3 - Divisibility by 5
            userGuess = int(input("Your guess: "))
            if userGuess == compNum :
                print("Awesome guess. You win!")
                break
            elif compNum % 5 == 0:
                print("Clue 3: The number is divisible by 5.")
                counter -= 1    
            else :
                print("Clue 3: The number is not divisible by 5.")  
                counter -= 1    

            # Last Guess
            userGuess = int(input("Your guess: "))
            if userGuess == compNum :
                print("That was a close call, but the right one! You win!")
                break
            else :
                print("Game over. The answer was " + str(compNum) + ".")
            
    else : 
        return
        
                    
```

```python
print("Time to play! We'll give you a hint and you guess from there. You have 3 chances. Good luck!")

# A Hint to Get Started
if compNum < 16 :
    print("Hint: The number is between 10 and 15, inclusively.")
else:
    print("Hint: The number is between 16 and 20, inclusively.")


userGuess = int(input("Your guess: "))


if userGuess == compNum :
    print("Are you a magician?! You win!")
else :
    guessNum(userGuess)
```
